```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Datasets
df = pd.read_csv("path/featured_cpi.csv", parse_dates=["Date"])
forecast_df = pd.read_csv("F:/ML/CPI/Weighted_CPI_Forecasting/cpi_forecast_6months.csv", parse_dates=["Date"])
```

Plot the historical and forecasted CPI together:

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x="Date", y="Weighted_CPI", label="Historical")
sns.lineplot(data=forecast_df, x="Date", y="Predicted_Weighted_CPI", label="Forecast", linestyle="--")
plt.title("All-India Weighted CPI: Historical & Forecast")
plt.xlabel("Date")
plt.ylabel("CPI")
plt.legend()
plt.tight_layout()
plt.show()
```

Display the predicted CPI for the next 6 months:

```python
forecast_df[["Date", "Predicted_Weighted_CPI"]]
```

**Inference**
- CPI has shown an upward trend with clear seasonal patterns.
- Forecasted CPI for the next 6 months continues the current trend.
- The model shows good predictive performance (RMSE  0.30).

**Conclusion**
1. Reliable Forecasting Achieved:
Using feature engineering and regression modeling, the next 6 months of weighted CPI were successfully forecasted with reasonable error margins (Test RMSE  0.30, MAE  0.25).

4. Subgroup Forecasting Planned:
The project is now transitioning to subgroup-level CPI predictions to better understand the components driving CPI and to forecast category-specific inflation trends.
